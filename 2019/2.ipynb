```python
import pandas as pd
```

```python
values = ""
with open('./data/2.txt') as f:
    values = f.readlines()[0][:-1].split(',')
    
values = list(map(int, values))
```

```python
def run_commands(values, noun=12, verb=2):
    values = values.copy()
    i = 0

    values[1] = noun
    values[2] = verb

    while i < len(values):
        command = values[i]
        
        if command == 99:
            break
        elif command == 1:
            values[values[i+3]] = values[values[i+1]] + values[values[i+2]] 
        elif command == 2:
            values[values[i+3]] = values[values[i+1]] * values[values[i+2]] 

        i += 4
    
    return values
```

```python
print(run_commands(values))
```

```python
for i in range(100):
    for j in range(100):
        if run_commands(values, i, j)[0] == 19690720:
            print(100 * i + j)
            
```
