```python
def line_points(route):
    current_point = [0, 0]
    points = set([])
    points = {}
    steps = 0
    
    for r in route:
        direction, distance = r[0], int(r[1:])
        
        if direction == 'L':
            for i in range(1, distance+1):
                current_point[0] -= 1
                steps += 1
                if tuple(current_point) not in points:
                    points[tuple(current_point)] = steps
        elif direction == 'R':
            for i in range(1, distance+1):
                current_point[0] += 1
                steps += 1
                if tuple(current_point) not in points:
                    points[tuple(current_point)] = steps
        elif direction == 'U':
            for i in range(1, distance+1):
                current_point[1] += 1
                steps += 1
                if tuple(current_point) not in points:
                    points[tuple(current_point)] = steps
        elif direction == 'D':
            for i in range(1, distance+1):
                current_point[1] -= 1
                steps += 1
                if tuple(current_point) not in points:
                    points[tuple(current_point)] = steps
     
    return points


def min_distance(line_1, line_2):
    line_1 = set(line_1.keys())
    line_2 = set(line_2.keys())
    return min([abs(x[0]) + abs(x[1]) for x in line_1.intersection(line_2)])


def intersection_steps(line_1, line_2):
    line_1_set = set(line_1.keys())
    line_2_set = set(line_2.keys())
    intersections = line_1_set.intersection(line_2_set)
    return min(line_1[intersection] + line_2[intersection] for intersection in intersections)
    
```

```python
line_1 = line_points(['R8','U5','L5','D3'])
line_2 = line_points(['U7','R6','D4','L4'])
print(min_distance(line_1, line_2))
print(intersection_steps(line_1, line_2))
```
