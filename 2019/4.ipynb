```python

    It is a six-digit number.
    The value is within the range given in your puzzle input.
    Two adjacent digits are the same (like 22 in 122345).
    Going from left to right, the digits never decrease; they only ever increase or stay the same (like 111123 or 135679).

Other than the range rule, the following are true:

    111111 meets these criteria (double 11, never decreases).
    223450 does not meet these criteria (decreasing pair of digits 50).
    123789 does not meet these criteria (no double).

How many different passwords within the range given in your puzzle input meet these criteria?

Your puzzle input is 353096-843212.
```

```python
def rule_adjacent_digits(guess):
    for i in range(len(guess)-1):
        if guess[i] == guess[i+1]:
            return True
    return False
```

```python
def rule_never_decreases(guess):
    for i in range(len(guess)-1):
        if guess[i] > guess[i+1]:
            return False
    
    return True
```

```python
def rule_adjacent_pairs(guess):
    for i in range(len(guess)-3):
        if guess[i] == guess[i+1] == guess[i+2] and guess[i] != guess[i+3]:
            return False
        
#     if len(set(guess[-3:])) == 1 and guess[-3] != guess[-4]:
#         return False
    
    return True
```

```python

def check_two_consecutive_digits(n):
    repeat_count = 0
    for n1, n2 in zip(n, n[1:]):
        if n1 == n2:
            repeat_count += 1
        else:
            if repeat_count == 1:
                return True
            repeat_count = 0
    return repeat_count == 1
```

```python
start = 353096
end = 843212
count = 0
for i in range(start,end+1):
    j = list(map(int, list(str(i))))
    if rule_adjacent_digits(j) and rule_never_decreases(j):
        count += 1
        
count
```

```python
start = 353096
end = 843212
count = 0
for i in range(start,end+1):
    j = list(map(int, list(str(i))))
    if rule_adjacent_digits(j) and rule_never_decreases(j) and check_two_consecutive_digits(j):
        count += 1
        
count
```
