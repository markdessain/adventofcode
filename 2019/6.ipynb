```python
values = []
with open('./data/6.txt') as f:
    for l in f.readlines():
        values.append(l[:-1])
```

```python

class Planet:
    
    def __init__(self, name):
        self.name = name
        self.planet = None
        
    def orbits(self, planet):
        self.planet = planet
        

def create_list(values):
    planets = {}
    
    for value in values:
        a, b = value.split(')')
        
        if a not in planets:
            planets[a] = Planet(a)
            
        if b not in planets:
            planets[b] = Planet(b)
            
        planets[b].orbits(planets[a])
    
    return planets
```

```python
i = 0
for planet in create_list(values).values():
    
    while True:
        if not planet.planet:
            break
        else:
            planet = planet.planet
            i += 1
            
print(i)
```

```python
planets = create_list(values)
```

```python
planet = planets['YOU']

trace_you = {}
i = 0
while True:
    if not planet.planet:
        break
    else:
        planet = planet.planet
        trace_you[planet] = i
        i += 1
```

```python
planet = planets['SAN']

trace_san = {}
i = 0
while True:
    if not planet.planet:
        break
    else:
        planet = planet.planet
        trace_san[planet] = i
        i += 1
```

```python
links = set(trace_you.keys()).intersection(set(trace_san.keys()))
```

```python
for trace, count in trace_you.items():
    if trace in links:
        print(count)
        break
```

```python
for trace, count in trace_san.items():
    if trace in links:
        print(count)
        break
```
